Photo Restoration using Inpainting

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('hasarlÄ±.jpg')
cv2_imshow(image)
cv2.waitKey(0)
```

```python
marked_damages = cv2.imread('mask.jpg', 0)
cv2_imshow(marked_damages)
cv2.waitKey(0)
```

```python
ret, thresh1 = cv2.threshold(marked_damages, 254,255, cv2.THRESH_BINARY)
cv2_imshow(thresh1)
cv2.waitKey(0)
```

```python
kernel = np.ones((7,7), np.uint8)
mask = cv2.dilate(thresh1, kernel, iterations = 1)
cv2_imshow(mask) #dilated mask
cv2.imwrite("sonuc.jpg", mask)
cv2.waitKey(0)

restored = cv2.inpaint(image, mask, 3, cv2.INPAINT_TELEA)
cv2_imshow(restored)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
